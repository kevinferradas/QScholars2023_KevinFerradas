### Problema 1

Solución: https://github.com/kevinferradas/QScholars2023_KevinFerradas/blob/main/prob1-min.png

```python
En quiskit:
```

```python
from qiskit.visualization import array_to_latex
from qiskit.quantum_info import random_statevector, Statevector 
import numpy as np
```

```python
from qiskit.quantum_info.operators import Operator
```

```python
from qiskit.visualization import plot_bloch_vector
```

```python
from qiskit.visualization import plot_state_qsphere
```

```python
from qiskit import QuantumCircuit
```

```python

ket_plus = Statevector.from_label('+')
ket_plus.draw('Latex')
```

```python
ket_minus = Statevector.from_label('-')
ket_minus.draw('Latex')
```

```python
ket_plus = [[1/np.sqrt(2)], [1/np.sqrt(2)]]
array_to_latex(ket_plus)
```

```python
ket_minus = [[1/np.sqrt(2)], [-1/np.sqrt(2)]]
array_to_latex(ket_minus)
```

```python
Operador_ket_plus = Operator(ket_plus)
Operador_ket_plus
```

```python
Operador_ket_minus = Operator(ket_minus)
Operador_ket_minus
```

```python
vector_estado1 = Statevector(Operador_ket_plus.tensor(Operador_ket_plus))
Statevector(Operador_ket_plus.tensor(Operador_ket_plus)).draw('Latex')
```

```python
vector_estado = Statevector(vector_estado1.tensor(Operador_ket_minus))
Statevector(vector_estado1.tensor(Operador_ket_minus)).draw('Latex')
```

Las probabilidades parciales para el qubit central son:

```python
probs_vector_estado = vector_estado.probabilities_dict([1])
print('probs: {}'.format(probs_vector_estado))
```
